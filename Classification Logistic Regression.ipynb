# Logistic Regression

Linear Regression is suited to estimate continuous values, Logistic Regression is suited for predict the class, the most probable class, the result will be a probability.

#### Read data (The data of a telco company's client and whether the customer left or not)

```python
import pandas as pd
import pylab as pl
import numpy as np
import scipy.optimize as opt
from sklearn import preprocessing
%matplotlib inline 
import matplotlib.pyplot as plt
```

```python
!wget -O ChurnData.csv https://s3-api.us-geo.objectstorage.softlayer.net/cf-courses-data/CognitiveClass/ML0101ENv3/labs/ChurnData.csv
```

```python
churn_df = pd.read_csv("ChurnData.csv")
churn_df.head()
```

#### Prepare data

```python
churn_df = churn_df[['tenure', 'age', 'address', 'income', 'ed', 'employ', 'equip',   'callcard', 'wireless','churn']]
churn_df['churn'] = churn_df['churn'].astype('int')
churn_df.head()
```

```python
X = np.asarray(churn_df[['tenure', 'age', 'address', 'income', 'ed', 'employ', 'equip']])
X[0:5]
```

```python
y = np.asarray(churn_df['churn'])
y [0:5]
```

```python
from sklearn import preprocessing
X = preprocessing.StandardScaler().fit(X).transform(X)
X[0:5]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split( X, y, test_size=0.2, random_state=4)
```

#### Modeling

```python
from sklearn.linear_model import LogisticRegression
LR = LogisticRegression(C=0.01, solver='liblinear').fit(X_train,y_train)
LR
```

```python
yhat = LR.predict(X_test)
yhat
```

```python
yhat_prob = LR.predict_proba(X_test)
yhat_prob
```

#### Evaluation

```python
from sklearn.metrics import jaccard_similarity_score
jaccard_similarity_score(y_test, yhat)
```

```python
from sklearn.metrics import log_loss
log_loss(y_test, yhat_prob)
```

Confusion matrix is also a good evaluation method
