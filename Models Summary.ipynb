# Models Summary

Regression -> number target variable
- Linear
- Multilinear
- Polynomial
- Non-Linear

Classification -> decision between seperate classes
- K-nearest neighbour
- Decision Trees
- Logistic Regression
- SVM (Support Vector Machine)

Clustering -> unsupervised
- K-Means
- Hiearchial
- DB-Scan

Evalution
- Regression -> R-square
- Classification -> Jaccard-score, Logloss, Confusion-matrix
- Clustering -> difficult because it is unsupervised, proper visualization can be useful
